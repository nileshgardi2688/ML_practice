Importing the dependecies

```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data collection and processing

```python
#loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
#loading data to pandas data from
data_frame= pd.DataFrame(breast_cancer_dataset.data, columns=breast_cancer_dataset.feature_names)
```

```python
#print the first 5 rows
data_frame.head()
```

```python
#adding the target column ro the data frome
data_frame['label'] = breast_cancer_dataset.target
```

Separating the features and target

```python
x = data_frame.drop(columns='label', axis=1)
y=data_frame['label']
```

```python
print(x)
```

```python
print(y)
```

#Splitting the data into traing and test data

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=2)
```

```python
print(x_train)
```

```python
print(x_test)
```

```python
print(y_test)
```

```python
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

Model training using logistic regression

```python
model= LogisticRegression()
```

```python
#training logostic regression model using training data
model.fit(x_train,y_train)
```

Model evaluation using accuracy score

```python
# accuracy on training data
x_train_prediction = model.predict(x_train)
```

```python
print(x_train_prediction)
```

```python
training_data_accuracy= accuracy_score(y_train, x_train_prediction)
```

```python
print("accuracy on training data= ", training_data_accuracy)
```

```python
#accuracy on test data
x_test_prediction = model.predict(x_test)
test_data_accuracy= accuracy_score (y_test, x_test_prediction)
```

```python
print("Accuracy on test data: ", test_data_accuracy)
```

Building a predictive sysyem

```python
input_data = (18.25,19.98,119.6,1040,0.09463,0.109,0.1127,0.074,0.1794,0.05742,0.4467,0.7732,3.18,53.91,0.004314,0.01382,0.02254,0.01039,0.01369,0.002179,22.88,27.66,153.2,1606,0.1442,0.2576,0.3784,0.1932,0.3063,0.08368)

#change the input array as numpy array
input_data_as_numpy_array = np.asarray(input_data)
```

```python
print(input_data_as_numpy_array)
```

```python
#reshape the numpy array as we are predicting one datapoint
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)
print(input_data_reshaped)
```

```python
prediction = model.predict(input_data_reshaped)
print(prediction)

if (prediction[0] == 0):
  print("Malignant")

else:
  print("Benign")
```
